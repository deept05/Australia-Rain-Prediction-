```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('/kaggle/input/weather-dataset-rattle-package/weatherAUS.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.drop(['Evaporation','Sunshine','Cloud9am','Cloud3pm','Location','Date'],axis = 1)
df.head()
```

```python
df = df.dropna(axis = 0)
df.shape
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
df['WindGustDir'] = le.fit_transform(df['WindGustDir'])
df['WindDir9am'] = le.fit_transform(df['WindDir9am'])
df['WindDir3pm'] = le.fit_transform(df['WindDir3pm'])
df['RainToday'] = le.fit_transform(df['RainToday'])
df['RainTomorrow'] = le.fit_transform(df['RainTomorrow'])
```

```python
df.head()
```

```python
x = df.drop(['RainTomorrow'],axis = 1)
y = df['RainTomorrow']
```

```python
x.head()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x = df['MaxTemp'], y = df['MinTemp'],hue = df['RainTomorrow'],palette = 'inferno')
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x = df['Humidity9am'], y = df['Temp9am'],hue = df['RainTomorrow'],palette = 'inferno')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot = True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
predictions = lr.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
predictions = dt.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
predictions = rf.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

XGBOOST Classifier

```python
import xgboost as xgb
xgb = xgb.XGBClassifier()
xgb.fit(x_train, y_train)
predictions = xgb.predict(x_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
print(accuracy_score(y_test, predictions))
```
